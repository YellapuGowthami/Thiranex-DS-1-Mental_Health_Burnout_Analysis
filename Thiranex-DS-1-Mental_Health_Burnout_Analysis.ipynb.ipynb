```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Dataset
df = pd.read_csv("/content/mental_health_burnout_tech_2026.csv")
```

```python
# Display First 5 Rows
print("First 5 Rows:")
print(df.head())
```

```python
# Display Last 5 Rows
print("Last 5 Rows:")
print(df.tail())
```

```python
# Dataset Information
print("\nDataset Information:")
print(df.info())
```

```python
# Missing Values Check
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Remove Duplicates
df.drop_duplicates(inplace=True)
```

```python
print("\nDuplicates Removed Successfully")
```

```python
# Dataset Shape
print("\nDataset Shape:")
print(df.shape)
```

```python
# -------------------------------
# Visualization 1
# Burnout Score Distribution
# -------------------------------

plt.figure(figsize=(8,5))
sns.histplot(df['burnout_score'], bins=20)

plt.title("Burnout Score Distribution")
plt.xlabel("Burnout Score")
plt.ylabel("Count")

plt.show()
```

```python
# -------------------------------
# Visualization 2
# Work Mode and Burnout
# -------------------------------

plt.figure(figsize=(8,5))

sns.countplot(
    x='work_mode',
    hue='burnout_level',
    data=df
)

plt.title("Work Mode and Burnout")

plt.show()
```

```python
# -------------------------------
# Visualization 3
# Stress vs Sleep
# -------------------------------

plt.figure(figsize=(8,5))

sns.scatterplot(
    x='sleep_hours_per_night',
    y='stress_score',
    data=df
)

plt.title("Sleep Hours vs Stress Score")

plt.show()
```

```python
# -------------------------------
# Visualization 4
# Correlation Heatmap
# -------------------------------

plt.figure(figsize=(14,10))

sns.heatmap(
    df.corr(numeric_only=True),
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.show()
```

Key Insights:

1. Employees with longer working hours tend to experience higher burnout.

2. Employees with less sleep show higher stress scores.

3. Work mode influences burnout and job satisfaction.

4. Better work-life balance reduces stress and burnout.

5. Mental health support helps improve employee well-being.
